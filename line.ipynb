```ruby
require 'nyaplot'
```

```ruby
path = File.expand_path("../data/11-1-result0.csv", __FILE__)
df = Nyaplot::DataFrame.from_csv(path)
```

```ruby
Nyaplot.init_iruby
```

```ruby
i=0
plot1 = Nyaplot::Plot.new
plot2 = Nyaplot::Plot.new
plot3 = Nyaplot::Plot.new
while i < 2
  path = File.expand_path("../data/11-1-result0." + i.to_s + ".csv", __FILE__)
  df = Nyaplot::DataFrame.from_csv(path)
  plot1.add(:line, df['t'], df['v'])
  plot2.add(:line, df['t'], df['w'])
  plot3.add(:line, df['v'], df['w'])
  i += 1
end
plot1.zoom(true)
plot2.zoom(true)
plot3.zoom(true)

frame = Nyaplot::Frame.new
frame.add(plot1)
frame.add(plot2)
frame.add(plot3)
frame.show
```
